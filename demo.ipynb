```python
iris = datasets.load_iris()
```

```python
iris_df = pd.DataFrame(columns=iris.feature_names, data=iris.data)
```

#### Датагаа `train`, `test` гэж хуваая 

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state = 30)
```

```python
y_train
```

#### Дата 80% нь `train`, 20% нь `test` д хуваагдсан байна.

```python
print(len(x_train))
print(len(x_test))
```

#### Модел сургахад `decision tree` ашиглая

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier(random_state=0)
```

#### [cross validation](https://en.wikipedia.org/wiki/Cross-validation_(statistics))  аар моделийн чанарыг шалгая

```python
from sklearn.model_selection import cross_val_score
```

```python
val = cross_val_score(clf, x_train, y_train, cv=10)
```

```python
val.mean()
```

#### `decision tree classifier` модел сургая. Дата нь жижигхэн тул модел хурдан үүснэ.

```python
clf.fit(x_train, y_train)
```

#### Сургасан моделоо ашиглаж тест дата дээрээ predict хийе

```python
y_pred = clf.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

#### `y_pred` нь тест датаг predict хийсэн утга агуулна. Predict хийсэн утгаа жинхэнэ утгатай нв харьцуулъя 

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred) * 100
```

#### 96.7% ийн нарийвлалтай зөв таажээ. `confusion matrix` аар юун дээр буруу таасныг харъя

```python
from sklearn.metrics import confusion_matrix
```

```python
pd.DataFrame(confusion_matrix(y_test, y_pred, labels=[0,1,2]), columns = ['setosa', 'versicolor', 'virginica'], index=['setosa', 'versicolor', 'virginica'])
```
